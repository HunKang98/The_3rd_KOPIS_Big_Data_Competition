```python
import pandas as pd
import numpy as np
import os
```

```python
# 각 년도별로 같은 작업을 진행하면 됩니다.
dirc = r'../provided_data/22년/'
ext = r'.csv'
```

```python
files_to_read = [file for file in os.listdir(dirc) if file.endswith(ext)]
```

예술의 전당에 해당하는 공연 추출하기

```python
aug_df = pd.DataFrame()
for f in files_to_read :
    file_path = os.path.join(dirc,f)
    df = pd.read_csv(file_path,encoding = 'cp949')
    df = df[df['공연시설코드'] == 'FC222223']
    df = df[df['공연장코드'] == 'FC222223-02']
    df.drop(['공연시설코드', '시설특성', '개관연도', '주소', '편의시설_레스토랑 여부', '편의시설_카페 여부',
       '편의시설_편의점 여부', '편의시설_놀이방 여부', '편의시설_수유실 여부', '장애인시설_주차장 여부',
       '장애인시설_화장실 여부', '장애인시설_경사로 여부', '장애인시설_전용엘리베이터 여부', '주차시설_자체 여부',
       '주차시설_공영 여부', '공연장코드', '장애인석', '무대시설_오케스트라피트 여부', '무대시설_연습실 여부',
       '무대시설_분장실 여부', '무대시설_무대넓이', '입장권고유번호',
       '예매/취소일시', '예매/취소방식코드',
       '예매/취소방식명(관리시스템)', '예매/취소방식명(전송처)', '결제수단코드', '결제수단명(관리시스템)',
       '결제수단명(전송처)', '할인금액', '할인종류코드', '할인종류명(관리시스템)', '할인종류명(전송처)',
       '성별', '연령', '공연지역명', '판매시작일시', '판매종료일시', '판매페이지URL'], axis = 1, inplace = True)

    aug_df = pd.concat([aug_df, df])
```

```python
aug_df.to_csv('../preprocessed_data/예술의전당/예술의전당_02_22년.csv',index = False, encoding='cp949')  
```

예술의 전당 출연진 추출하기

```python
hall02_ = aug_df.drop_duplicates(subset=['공연코드'])
hall02_
```

```python
trend_index = pd.DataFrame({'공연코드': hall02_['공연코드'],
                         '출연진내용': hall02_['출연진내용']})
```

```python
trend_index
```

```python
trend_index.to_csv('../trend_index/예술의전당/예술의전당 출연진_22년1.csv',index = False, encoding='cp949')
```
